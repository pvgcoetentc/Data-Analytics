1.1 Understanding the Data

The data set is the backbone of the problem at hand. Choosing the right kind of dataset with right amount of data is very crucial in solving any data science problem. Some of the basic requirements that the dataset should satisfy are:

a. Relevant Data:
Remember you cannot use any dataset to solve a particular problem. So what does a ‘relevant’ dataset looks like?
image.png

In the examples above, the left table dataset columns are totally not related to each other. The patient identifier has nothing to do with the current stock price and the premium amount of different insurance plans.
Whereas, in the table on right, each row can be associated with a particular patient. The different columns are actually qualifying the patient and are termed as ‘Features’ of the dataset.

b. Size of data:

Estimating the right amount of data needed is crucial while solving a data science problem. The rule of thumb is for e.g. if you have say 20 categorical variables with 15 levels / classes each and 18 continuous variables (for continuous variables assume 10 levels / classes) then the minimum number of rows needed is: (20 * 15 + 18 * 10) * 100 = 48,000 rows. (100 is just a multiplying factor and can be varied depending on how much data is available and the application area under consideration)
Also, there is a very important concept of ‘sparse data’. If the dataset has a lot of missing values then the data is called a sparse data. There are various ways (like KNN imputation, Average / Median value imputation etc.) in which we can handle the missing values depending on the application area.

1.4  **Getting Started Analyzing Data in Python**

Creating A DataFrame in Pandas::   Creation of dataframe is done by passing multiple Series into the DataFrame class using pd.Series method. Here, it is passed in the two Series objects, s1 as the first row, and s2 as the second row.
Example:

```python
# assigning two series to s1 and s2
s1 = pd.Series([1,2])
s2 = pd.Series(["Ashish", "Sid"])
# framing series objects into data
df = pd.DataFrame([s1,s2])
# show the data frame
df
  
```

```python
# data framing in another way
# taking index and column values
dframe = pd.DataFrame([[1,2],["Ashish", "Sid"]],index=["C1", "C2"],columns=["R1", "R2"])
dframe
```

```python
# framing in another way 
# dict-like container
dframe = pd.DataFrame({
        "c1": [1, "Ashish"],
        "c2": [2, "Sid"]})
dframe
```

Scatter Plots and Bar Graphs

```python
sct = np.random.rand(20, 2)
print (sct)
plt.scatter(sct[:,0], sct[:,1]) # I am plotting a scatter plot.
```

```python
ghj =[5, 10 ,15, 20, 25]
it =[ 1, 2, 3, 4, 5]
plt.bar(ghj, it) # simple bar graph
```

```python
ghj =[5, 10 ,15, 20, 25]
it =[ 1, 2, 3, 4, 5]
plt.bar(ghj, it, width =5)# you can change the thickness of a bar, by default the bar will have a thickness of 0.8 units
```

```python
ghj =[5, 10 ,15, 20, 25]
it =[ 1, 2, 3, 4, 5]
plt.barh(ghj, it) # barh is a horizontal bar graph
```

Multiple bar charts

```python
new_list = [[5., 25., 50., 20.], [4., 23., 51., 17.], [6., 22., 52., 19.]]
x = np.arange(4) 
plt.bar(x + 0.00, new_list[0], color ='b', width =0.25)
plt.bar(x + 0.25, new_list[1], color ='r', width =0.25)
plt.bar(x + 0.50, new_list[2], color ='g', width =0.25)
```

```python
#Stacked Bar charts
p = [5., 30., 45., 22.]
q = [5., 25., 50., 20.]
x =range(4)
plt.bar(x, p, color ='b')
plt.bar(x, q, color ='y', bottom =p) 
```

If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.
