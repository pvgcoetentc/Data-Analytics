### 3.6 Binning: 

- Binning is used for data smoothing and also to remove noisy data, this is done by converting continuous variables into categorical variables. In this method the data is initially sorted and then the sorted values are distributed into different bins. There are mainly two
types in unsupervised binning:
a)Equal Frequency binning
b)Equal Width binning

- a)Equal Frequency binning : 
This type of binning divides the data into bins such that every
bin will be of same height i.e it will distribute the values of the data equally into formed
categories.Using the below formula we can calculate the frequency,

Freq=n/x

where n = number of values in data
x = number of categories
Freq = frequency of the category

- b)Equal Width binning : 
This type of binning divides the data into bins such that every bin will be of same width i.e it divides the continuous variable into several categories of the same range.Using the below formula we can calculate the width of the bin,

W=(max-min) / x

where max = maximum value in the list
min = minimum value in the list
x = number of categories
W = width of category

### 3.7 Turning Categorical variables into quantitative variables : 
- As the machine learning models cannot work on the categorical variables they changed into quantitative form. Using the “get_dummies()” function we can perform the task of converting the categorical variables into boolean values.

```python
data = {'name': ['Monica', 'Rachel', 'Pheobe', 'Chandler', 'Joey', 'Ross'],
                'like_music': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
                'gender': ['female', 'female', 'female', 'male', 'male', 'male']}

df = pd.DataFrame(data, columns = ['name','like_music', 'gender'])
print(df)
```

```python
df_music = pd.get_dummies(df['like_music'])
df_music
```

```python
df_new = pd.concat([df, df_music], axis=1)
df_new
```
