## **4.1 Probablity**

### 4.1.1 Uniform Distribution

Let us generate 50,000 data points from a uniform distribution spanning the range 0 to 10. In the density plot, we see that the density of our uniform data is essentially level meaning any given value has the same probability of occurring. The area under a probability density curve is always equal to 1.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import random as rdm
```

stats.distribution.rvs() generates random numbers from the specified distribution. The arguments to rvs() will vary depending on the type of distribution you're working with; in the case of the uniform distribution, we have to specify the starting and ending points and the size.

```python
uniform_data = stats.uniform.rvs(size=50000, loc=0, scale=10)
#This generates 50000 random numbers from 0 to 10
```

```python
pd.DataFrame(uniform_data).plot(kind="density",figsize=(9,9),xlim=(-1,11))
```

stats.distribution.pdf() gives you the probability density (height of the distribution) at a given x value. Since the uniform distribution is flat, all x values within its range will have the same probability density and x values outside the range have a probability density of 0.

```python
for x in range(-1,12,1):
    print("Density at x = " + str(x))
    print( stats.uniform.pdf(x, loc=0, scale=10) )
    #pdf for all numbers from -1 to 11
```

stats.distribution.cdf() is used to determine the probability that an observation drawn from a distribution falls below a specified value. cdf() gives you the area under the distribution's density curve to the left of a certain value on the x axis.

```python
stats.uniform.cdf(x=4,loc=0,scale=10)
#probability of getting a number below 4 in a trial
```

```python
stats.uniform.cdf(x=8,loc=0,scale=10)
#probability of getting a number below 8 in a trial
```

stats.distribution.ppf() is the inverse of cdf(): it returns the x axis cutoff value (quantile) associated with a given probability.

```python
stats.uniform.ppf(q=0.4,loc=0,scale=10)
#cutoff value on X-axis for probability 0.4
```

```python
stats.uniform.ppf(q=0.8,loc=0,scale=10)
#cutoff value on X-axis for probability 0.8
```

### 4.1.2 Binomial Distribution


Let us model flipping a fair coin 100000 times with a binomial distribution where the number of trials is set to 10 and the probability of success is set to 0.5. In this case the distribution would tell us how likely it is to get zero heads, 1 head, 2 heads and so on. We will perform the experiment for 10000 trials.
The name for the binomial is binom. Let's generate and investigate some binomial data:



```python
#creation of dataset
fair_coin_flips = stats.binom.rvs(n=10,p=0.5,size=10000)
# n is number of flips per trial, p is the probabiltiy of success and size is the number of trials
print( pd.crosstab(index="counts", columns= fair_coin_flips))
#displaying the number of successful trials for different number of heads
```

```python
pd.DataFrame(fair_coin_flips).hist(range=(-0.5,10.5), bins=11)
```

Cumilative Distribution Function:
For continuous probability density functions, the cdf() function lets us check the probability of achieving a number of successes within a certain range.

```python
stats.binom.cdf(k=5,n=10,p=0.5) 
#probability of 5 successes or less from a trial 
```

```python
stats.binom.cdf(k=7,n=10,p=0.5) 
#probability of 7 successes or less from a trial
```

Probability Mass Function:
For discrete distributions like the binomial, use pmf() (probability mass function) to check the mass (proportion of observations) at given number of successes k.

```python
stats.binom.pmf(k=5,n=10,p=0.5) 
#probability of exact 5 successes from a trial
```

```python
stats.binom.pmf(k=7,n=10,p=0.5)
#probability of exact 5 successes from a trial
```

### 4.1.3 Poisson Distribution

The Poisson distribution models the probability of seeing a certain number of successes within a time interval, where the time it takes for the next success is modeled by an exponential distribution.
Let us model a poisson distribution for the sale of mobiles for a mobile shop where the average sale is 2 mobiles per hour. We will perform the experiment for 10000 hours.
The name for Poisson Distribution is 'poisson'.

```python

rdm.seed(12)
#creation of dataset
sale = stats.poisson.rvs(size=10000, mu=2)
#mu is the average sale per hour and we will create a dataset for 10000 hours
print( pd.crosstab(index="counts", columns= sale))
#displaying the number of mobiles sold for various number of hours
```

```python
pd.DataFrame(sale).hist(range=(-0.5,max(sale)+0.5),bins=max(sale)+1)
```

We can use cdf() to check the probability of achieving more or less than a certain number of successes. 

```python
stats.poisson.cdf(k=2,mu=2) 
#probability of 2 or less mobiles sold at any random hour
```

```python
stats.poisson.cdf(k=4,mu=2) 
#probability of 4 or less mobiles sold at any random hour
```

We can use pmf() to check the probability of obtaining a specific number of successes.

```python
stats.poisson.pmf(k=2,mu=2) 
#probability of exact 2 mobiles sold at any random hour
```

```python
stats.poisson.pmf(k=4,mu=2) 
#probability of exact 4 mobiles sold at any random hour
```
