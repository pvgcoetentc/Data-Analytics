##**4.3 Hypothesis Testing (One sampled T-test)**

###**4.3.1 Basics** 

 #### Coin Toss 
 Let us test that the probability of a coin toss to land on heads is 0.5
 

```python
#you want to test that a coin has exactly a 50% chance of landing on heads
import pandas as pd
```

```python
# data set for sample size of 50 coin flips 
file = (r'C:\Users\Windows\Desktop\suubmission\DataAnalytics\data set\coinFlip.csv')
data = pd.read_csv(file)
data.head(10)
```

```python
i = 0
Li = []
```

```python
#creating a sample space

while i < 1000:
    datatemp = data.sample(5)
    Esh = datatemp['End: H'].sum()
    Psh = Esh / 5
    Li.append(Psh)
    i = i + 1
```

```python
#calculating mean and variance

def Average(lst): 
    return sum(lst) / len(lst)
mean = Average(Li)
var = sum((i - mean) ** 2 for i in Li) / len(Li)
print('mean is :',mean)
print('variance is :',var)
```

```python
#normal distribution curve

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = mean
variance = var
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

### **4.3.2 Steps**

1. Formulate the Hypotheses
2. Select an appropriate Test
3. Choose Level of Significance, α
4. Collect Data and Calculate Test Statistic
5. Determine the Probability
6. Compare the probability (or Critical value) and make the decision

####**1. Hypothesis**

In this example we are testing that a coin has exactly a 50% chance of landing on heads 

The null hypothesis would be that 50% is correct, and the alternative hypothesis would be that 50% is not correct

Ho: u = 0.5  (NULL hypothesis)

H1: u != 0.5 (Alternate hypothesis)

#### 2. **Test Selection**

Different hypothesis tests use different test statistics based on the probability model assumed in the null hypothesis. 

In our case we will use T test 

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

We will use one sample T-test as it determines whether the sample mean is statistically different from a known or hypothesised population mean.


#### **3. Significance value**

It is necessary to balance the type I error and type II error. 

As a compromise, α is often set at 0.05; sometimes it is 0.01; other values of α are rare. We will consider 0.05 for our case

#### **4. Calculate Test Statistic**

Sample size is determined after taking into account the desired α and other qualitative considerations

In our case we are taking a sapmle size of 30

#### **5. Determine the Probability & 6. Compare the probability and make the decision**

Determine the Probability or the critical value

The probability associated with the calculated or observed value of the test statistic is p-value . This is the probability of getting a P-Value  when π=0.50. This is more than the level of significance of 0.05. Hence, the null hypothesis is accepted.

Golden rule - "If P(p-value) is Low, reject the Ho"


```python
from scipy.stats import ttest_1samp


datasamp = data.sample(30)
Eh = datasamp['End: H'].sum()
Ph = Eh / 20


print('probability of head in this sample is ', Ph)
tset, pval = ttest_1samp(datasamp['End: H'], 0.5)
print('p-values', pval)


if pval < 0.05:    # alpha value is 0.05 or 5%
    print(' we are rejecting null hypothesis')
else:
    print('we are accepting null hypothesis')
```
